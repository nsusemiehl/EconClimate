```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
directory = './/data//unemployment_IN' # government data

dfs_by_city = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if "Unemployment_Rate_M_Total" in f:
        continue
        
    df = pd.read_csv(f)
    dfs_by_city.append(df)
```

```python
all_unemployment_data = pd.concat(dfs_by_city)
all_unemployment_data["Date"] = pd.to_datetime(all_unemployment_data.Date).dt.date
all_unemployment_data = all_unemployment_data.rename(columns={"Region":"City", " Estimated Unemployment Rate (%)":"Estimated Unemployment Rate (%)"})
all_unemployment_data = all_unemployment_data.sort_values(["Date", "City"])
all_unemployment_data = all_unemployment_data[["Date", "City", "Estimated Unemployment Rate (%)"]]
# all_unemployment_data = all_unemployment_data.rename(columns={" Frequency":"Frequency", " Estimated Unemployment Rate (%)":"Estimated Unemployment Rate (%)", " Estimated Employed":"Estimated Employed", " Estimated Labour Participation Rate (%)":"Estimated Labour Participation Rate (%)"})
all_unemployment_data
```

```python
all_unemployment_data.isna().sum()
```

```python
print(min(all_unemployment_data["Date"]))
print(max(all_unemployment_data["Date"]))
```

```python
all_unemployment_data = all_unemployment_data[all_unemployment_data["Date"]>datetime.date(2018,1,1)]
```

```python
plt.hist(all_unemployment_data["Estimated Unemployment Rate (%)"], density=True, alpha=0.5)
plt.xlabel("Estimated Unemployment Rate (%)")
plt.show()

plt.boxplot(all_unemployment_data["Estimated Unemployment Rate (%)"])
plt.xlabel("Estimated Unemployment Rate (%)")
plt.show()
```

```python
all_unemployment_data_date = all_unemployment_data.groupby("Date").mean().reset_index()

plt.plot_date(all_unemployment_data_date["Date"], all_unemployment_data_date["Estimated Unemployment Rate (%)"])
plt.ylabel("Estimated Unemployment Rate (%)")
plt.show()

print(all_unemployment_data_date.nlargest(2, "Estimated Unemployment Rate (%)"))
```

```python
all_unemployment_data_city = all_unemployment_data.groupby("City").mean().reset_index()

plt.bar(all_unemployment_data_city["City"], all_unemployment_data_city["Estimated Unemployment Rate (%)"])
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=-45)
plt.show()
```
