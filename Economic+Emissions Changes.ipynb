```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from tabulate import tabulate
```

```python
econclimate = pd.read_csv(".//data//processed_new//econclimate.csv")
econclimate["m"] = econclimate["Y-m"].apply(lambda x: int(x.split("-")[1]))
econclimate["Y"] = econclimate["Y-m"].apply(lambda x: int(x.split("-")[0]))
econclimate
```

```python
quarters = []
for i in range(len(econclimate.index)):
    m = econclimate["m"].values[i]
    Y = econclimate["Y"].values[i]
    if 1 <= m <= 3:
        Q = 1
    elif 4 <= m <= 6:
        Q = 2
    elif 7 <= m <= 9:
        Q = 3
    elif 10 <= m <= 12:
        Q = 4
    
    quarters.append(f"{Y}-{Q}")
    
econclimate["Q"] = quarters  
econclimate
```

```python
scaler = MinMaxScaler()
scaler.fit(econclimate[['CO', 'NO2', 'SO2', 'PM2_5']])
```

```python
dfs = []
for city in econclimate["City"].unique():
    city_data = econclimate[econclimate["City"] == city]
    
    city_pollutants = scaler.transform(city_data[['CO', 'NO2', 'SO2', 'PM2_5']])
    city_data["Normalized AQI"] = np.mean(city_pollutants, axis=1)
    
    city_data = city_data[["Estimated Unemployment Rate (%)", "Normalized AQI", "Q"]].groupby("Q").mean().reset_index()
    city_data = pd.concat([city_data["Q"], city_data[['Estimated Unemployment Rate (%)', 'Normalized AQI']].diff()], axis=1).dropna()
    city_data = city_data.rename(columns={"Estimated Unemployment Rate (%)":"Change in Unemployment Rate", "Normalized AQI":"Change in Normalized AQI"})
    city_data["City"] = np.repeat(city, len(city_data.index))
    
    dfs.append(city_data)
```

```python
all_city_dfs = pd.concat(dfs)
quarter_dfs = []
for q in all_city_dfs["Q"].unique():
    quarter_data = all_city_dfs[all_city_dfs["Q"]==q].reset_index(drop=True)
    
    quads = []
    for i in range(len(quarter_data.index)):   
        if (quarter_data.iloc[i,1] >= 0) and (quarter_data.iloc[i,2] >= 0):
            quadrant = 1
        elif (quarter_data.iloc[i,1] >= 0) and (quarter_data.iloc[i,2] < 0):
            quadrant = 2
        elif (quarter_data.iloc[i,1] < 0) and (quarter_data.iloc[i,2] >= 0):
            quadrant = 3
        elif (quarter_data.iloc[i,1] < 0) and (quarter_data.iloc[i,2] < 0):
            quadrant = 4
        quads.append(quadrant)

    quarter_data["Quadrant"] = quads
#     quarter_data = quarter_data.rename(columns={"Estimated Unemployment Rate (%)":"Economic Change", 'Normalized AQI':'Emissions Change'})
#     quarter_data = quarter_data.drop(columns=['Estimated Unemployment Rate (%)', 'Normalized AQI'])
    quarter_dfs.append(quarter_data)
```

```python
all_change_data = pd.concat(quarter_dfs)
all_change_data
```

```python
for q in np.unique(all_change_data["Q"]):
    print(q)
    subset = all_change_data[all_change_data["Q"]==q].sort_values(by="Quadrant")
    
    matrix = np.empty((2, 2), dtype=object)
    for i in np.ndindex(matrix.shape): matrix[i] = []
    
    if 1 in subset["Quadrant"]:
        matrix[0,0].extend(subset[subset["Quadrant"]==1]["City"])
    if 2 in subset["Quadrant"]:
        matrix[0,1].extend(subset[subset["Quadrant"]==2]["City"])
    if 3 in subset["Quadrant"]:
        matrix[1,0].extend(subset[subset["Quadrant"]==3]["City"])
    if 4 in subset["Quadrant"]:
        matrix[1,1].extend(subset[subset["Quadrant"]==4]["City"])
        
    print(matrix[0,0])
    print(matrix[0,1])
    print(matrix[1,0])
    print(matrix[1,1])
    print()
```
